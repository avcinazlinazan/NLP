## Removing Punctuation and Numbers

w.isalpha(), tokenin sadece alfabetik karakterler (harfler) ierip iermediini kontrol eden bir metodur ve eer ieriyorsa True dndrr.
Eer w sadece harflerden oluuyorsa (yani herhangi bir noktalama iareti veya rakam iermiyorsa), bu token tokens_without_punc listesine dahil edilir.

w.isalnum(), tokenin yalnzca alfanmerik karakterleri (harf veya rakam) ierip iermediini kontrol eden bir metodur ve eer ieriyorsa True dndrr.
Eer w yalnzca harf veya rakamlardan oluuyorsa, bu token tokens_without_punc listesine dahil edilir.

## Cleaning Function - for classification (NOT for sentiment analysis)

```python
def cleaning(data):
    
    #1. Tokenize
    text_tokens = word_tokenize(data.lower()) 
    
    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    #4. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
    #joining
    return " ".join(text_cleaned)
```

## CountVectorization

```python
X = df2["product_Info"]
y = df2["productName"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 42)
```

 **stratify = y** parametresi kullanlm. Bu sayede y deikenindeki snf oranlar, hem eitim setinde hem de test setinde ayn ekilde korunmu olacak. Bu da modelin daha dengeli bir ekilde eitilmesine ve test edilmesine olanak salar.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()                      # ML DE SCALE DE YAPTIIMIZI GB YAPIYORUZ.
X_train_count = vectorizer.fit_transform(X_train)   # fit deyince btn unique leri tespit ediyor, transform da ise sayyor
X_test_count = vectorizer.transform(X_test)         # X testte olan kelime X train de yok ise deerlendirilmeye alnmaz.
                                                    # O yzden byk Corpus ta eitimesi gerekir ki iyi sonular alalm...
```

Burada verileri bir saym vektrne (bag of words) dntrmek iin CountVectorizer kullanyorsunuz. CountVectorizer, metin verilerini kelime seviyesinde saym vektrlerine dntrmek iin yaygn olarak kullanlan bir yntemdir.

Veri dnm ilemi iki aamada gerekleir:

fit_transform(X_train): Bu admda, eitim verileri (X_train) zerinde CountVectorizer' eitiyorsunuz. Bu ilem, tm benzersiz kelimeleri tanr ve her kelimenin bir stunda temsil edildii bir saym matrisini oluturur. Her hcrede bir kelimenin grlme skl bulunur.

transform(X_test): Bu admda, test verileri (X_test) zerinde ayn dnm uyguluyorsunuz. Ancak, burada sadece eitim verilerinde grdnz kelimelerin saysn sayyor ve bu kelime saylarn test verileri iin bir saym matrisine dntryorsunuz. Yani, eitim verileriyle renilen kelime daarcna dayal olarak test verilerini temsil eden bir saym matrisi oluturulur.

Sonu olarak, X_train_count ve X_test_count matrisleri, metin verilerini bag of words temsilinde saym vektrlerine dntren ilemi ifade eder. Bu temsil, metin verilerini saysal olarak ilemek ve makine renimi modellerine beslemek iin kullanlr.
